```python
# Data for the bar plot
gender = ['Male', 'Female']
index = [577, 314]
colors = ['#39ff14', '#F8FF14']
# Create the bar plot
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(gender, index, color=colors)
plt.xlabel("Gender")
plt.ylabel("No of people onboarding ship")
plt.title("Number of People Onboarding Ship by Gender")
plt.show()
```

```python
#Univariate Analysis:
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(data['Age'].dropna(), kde=True, color='darkblue')
plt.title('Age Distribution')
plt.subplot(1, 2, 2)
sns.countplot(data['Sex'], color='pink')
plt.title('Gender Distribution')
plt.show()
```

```python
# Data for the pie chart
labels = ['C Cherbourg', 'Q Queenstown', 'S Southampton']
sizes = [0.553571, 0.389610, 0.336957]
colors = ['#F83E76', '#F07EFF', '#14DCFF']
explode = (0.1, 0, 0) 
# Create a pie chart
fig = plt.figure()
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
# Plotting the pie chart with customizations
ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.2f%%', 
       shadow=True, startangle=140)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="Age", y="Fare", hue="Survived", palette='bright', s=50, edgecolor='k', alpha=0.7)
plt.title("Scatter Plot of Age vs Fare", fontsize=16)
plt.xlabel("Age", fontsize=14)
plt.ylabel("Fare", fontsize=14)
plt.legend(title="Survived", title_fontsize='13', fontsize='12', loc='upper right')
plt.grid(True)
plt.show()
```
